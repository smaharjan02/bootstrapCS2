```python
#plot the data
# Data for N=200

data_200 = [first_join_df_200[column_name].mean(), 
            second_join_df_200[column_name].mean(),
            third_join_df_200[column_name].mean(),
            fourth_join_df_200[column_name].mean()]
#print(data_200)
# Data for N=2000 
data_2000 = [first_join_df_2000[column_name].mean(), 
             second_join_df_2000[column_name].mean(),
             third_join_df_2000[column_name].mean(),
             fourth_join_df_2000[column_name].mean()]
#print(data_2000)

# Convert data_200 and data_2000 lists to DataFrames with index starting from 1
df_200 = pd.DataFrame(data_200, columns=['Data_200'], index=range(1, len(data_200) + 1))
df_2000 = pd.DataFrame(data_2000, columns=['Data_2000'], index=range(1, len(data_2000) + 1))

# Concatenate the two DataFrames horizontally (column-wise)
concatenated_df = pd.concat([df_200, df_2000], axis=1)

print(concatenated_df)

```
