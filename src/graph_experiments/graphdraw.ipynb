```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import csv
import numpy as np
import seaborn as sns
```

```python
#Read the file from first join
#Read the first csv file for comparison
df1 = pd.read_csv ('graph_experiment_results_query_first_join/graphlogs_log_tpch_1g.db_sr0.1_lt10_bs2000.csv')
#Read the Second csv file for comparison
df2 = pd.read_csv ('graph_experiment_results_query_first_join/graphlogs_log_tpch_1g.db_sr0.5_lt10_bs2000.csv')
#Read the third csv file for comparison
df3 = pd.read_csv ('graph_experiment_results_query_first_join/graphlogs_log_tpch_1g.db_sr1.0_lt10_bs2000.csv')
0
#Read the file for second join
#Read the first csv file for comparison
df4 = pd.read_csv ('graph_experiment_results_query_second_join/graphlogs_log_tpch_1g.db_sr0.1_lt10_bs2000.csv')
#Read the Second csv file for comparison
df5 = pd.read_csv ('graph_experiment_results_query_second_join/graphlogs_log_tpch_1g.db_sr0.5_lt10_bs2000.csv')
#Read the third csv file for comparison
df6 = pd.read_csv ('graph_experiment_results_query_second_join/graphlogs_log_tpch_1g.db_sr1.0_lt10_bs2000.csv')

#Read the file for third join
#Read the first csv file for comparison
df7 = pd.read_csv ('graph_experiment_results_query_third_join/graphlogs_log_tpch_1g.db_sr0.1_lt10_bs2000.csv')
#Read the Second csv file for comparison
df8 = pd.read_csv ('graph_experiment_results_query_third_join/graphlogs_log_tpch_1g.db_sr0.5_lt10_bs2000.csv')
#Read the third csv file for comparison
df9 = pd.read_csv ('graph_experiment_results_query_third_join/graphlogs_log_tpch_1g.db_sr1.0_lt10_bs2000.csv')

#Read the file for fourth join
#Read the first csv file for comparison
df10 = pd.read_csv ('graph_experiment_results_query_fourth_join/graphlogs_log_tpch_1g.db_sr0.1_lt10_bs2000.csv')
#Read the Second csv file for comparison
df11 = pd.read_csv ('graph_experiment_results_query_fourth_join/graphlogs_log_tpch_1g.db_sr0.5_lt10_bs2000.csv')
#Read the third csv file for comparison
df12 = pd.read_csv ('graph_experiment_results_query_fourth_join/graphlogs_log_tpch_1g.db_sr1.0_lt10_bs2000.csv')

```

```python
#Creating dataframes for each join_types (first,second,third,fouth)
# Concatenate all three sample for first join
first_join_df = pd.concat([df1.set_index("Query Number"),df2.set_index("Query Number"),df3.set_index("Query Number")],axis=1)
#print(first_join_df.head())

# Concatenate all three sample for second join
second_join_df = pd.concat([df4.set_index("Query Number"), 
                     df5.set_index("Query Number"), 
                     df6.set_index("Query Number")], axis=1)
#print(second_join_df.head())

# Concatenate all three sample for third join
third_join_df = pd.concat([df7.set_index("Query Number"), 
                     df8.set_index("Query Number"), 
                     df9.set_index("Query Number")], axis=1)
#print(third_join_df.head())

# Concatenate all three sample for fourth join
fourth_join_df = pd.concat([df10.set_index("Query Number"), 
                     df11.set_index("Query Number"), 
                     df12.set_index("Query Number")], axis=1)
print(fourth_join_df.head())
```

```python
#Hit percentage chart for first join dataframe
with PdfPages('Hit Percentage of 100m data with N = 200 for first join.pdf') as pdf:
   
    #print(multidf)
    g = first_join_df["Hit Percentage"].plot.bar( figsize=(6, 4))
    plt.xticks(rotation=0)
   
    h,l= g.get_legend_handles_labels()
    kw = dict(ncol=2, loc="lower center", frameon=False) 
    leg1 = g.legend(['0.1%', '0.5%'], bbox_to_anchor=[0.5,1.04],**kw)
    leg2 = g.legend(h[2:],['1%'], bbox_to_anchor=[0.5,1.00],**kw)
    g.add_artist(leg1)
    #plt.legend(['0.1%', '0.5%', '1%'],bbox_to_anchor=(.3, 1.08), loc='best')
    g.set_ylabel('Hit Percentage')
   
    #g.set_title('Hit Percentage of 100MB data with N = 200')
    
    pdf.savefig(bbox_inches='tight')
```

```python
# Bootstrap time taken chart for first join dataframe

with PdfPages('Bootstrap Time of 100m data with N = 200 for first join.pdf') as pdf:
   
    #print(multidf)
    g = first_join_df["Bootstrap Time"].plot.bar( figsize=(6, 4))
    plt.xticks(rotation=0)
   
    h,l= g.get_legend_handles_labels()
    kw = dict(ncol=2, loc="lower center", frameon=False) 
    leg1 = g.legend(['0.1%', '0.5%'], bbox_to_anchor=[0.5,1.08],**kw)
    leg2 = g.legend(h[2:],['1%'], bbox_to_anchor=[0.5,1.00],**kw)
    g.add_artist(leg1)
    #plt.legend(['0.1%', '0.5%', '1%'],bbox_to_anchor=(.3, 1.08), loc='best')
    g.set_ylabel('Bootstrap Time')
   
    #g.set_title('Hit Percentage of 100MB data with N = 200')
    
    pdf.savefig(bbox_inches='tight')
```

```python
#DataFrame for multiple join where sample size = 0.1
#compare Hit percentage of each query for each join
multi_join_so1_df = pd.concat([df1.set_index("Query Number"), 
                     df4.set_index("Query Number"),
                     df7.set_index("Query Number"), 
                     df10.set_index("Query Number")], axis=1)
print(multi_join_so1_df.tail())
```

```python
#Bar chart for multiple join where sample size = 0.1
with PdfPages('Bootstrap Time taken of 100m data with N = 2000 for sample 0.1 and multiple join.pdf') as pdf:
   
    g = multi_join_so1_df["Bootstrap Time"].plot.bar(figsize=(6, 4), 
                color=['blue', 'orange', 'green', 'red'])
    plt.xticks(rotation=0)

    # Get labels from plot
    handles, labels = g.get_legend_handles_labels()

    # Create single legend
    legend = g.legend(handles,
                    ['1st', '2nd', '3rd', '4th'],
                    labelcolor=['blue', 'orange', 'green', 'red'],
                    loc='upper left', #location
                    bbox_to_anchor=(1.05, 1), #position
                    title='Joins',
                    borderaxespad=0)     

    g.set_ylabel('Bootstrap Time')
    g.set_title("Bootstrap Time taken for 100m data with N = 200 for sample 0.1 and multiple join", fontsize=12, y=1.05)
    pdf.savefig(bbox_inches='tight')

```
